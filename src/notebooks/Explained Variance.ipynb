```python
%matplotlib nbagg

import matplotlib.pyplot as plt
import numpy as np

inp = np.load('../../data/explained_variance.npy')
```

```python
fig1 = plt.figure(figsize=(15,5))
ax1 = fig1.add_subplot(111)

ax1.bar(np.arange(150), inp[:150])
plt.title('DNA-Methylation Explained variance of the first 150 principal components')
plt.xlabel('n_component')
plt.ylabel('Explained Variance')
```

```python
in2 = np.load('../../data/protein_explained_variance.npy')
fig2 = plt.figure(figsize=(15,5))
ax2 = fig2.add_subplot(111)

ax2.bar(np.arange(150), in2[:150])
plt.title('Protein-Expression -- Explained variance of the first 150 principal components')
plt.xlabel('n_component')
plt.ylabel('Explained Variance')
```
