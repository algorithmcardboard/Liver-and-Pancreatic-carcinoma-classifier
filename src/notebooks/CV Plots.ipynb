```python
%matplotlib nbagg

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/grid_linear_cancer_search_scores_1450680262.out')
df
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
for c in [0.03125]:
    tmp = X[X[:,1] == c]
    print(tmp)
    plt.plot(tmp[:,2], tmp[:,0])
```

```python
X_axes = ['L_svm', 'G_svm', 'AdaBoost', 'Logistic Regression', 'G. Naive Bayes']
heights = [85.71, 93.51, 90.90, 92.21, 81.82]
colors = ['#C95834','#728CAC','#AA8D66','#319E98','#AF69AF']

ind = np.arange(len(X_axes))

fig2 = plt.figure(figsize=(12,5))
ax2 = fig2.add_subplot(111)
ax2.bar(ind+0.125, heights, width=0.25, color=colors)
plt.xticks(ind+0.25, X_axes, rotation=0)
```
