```python
%matplotlib nbagg

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/grid_linear_cancer_search_scores_1450680262.out')
```

```python
X = df.as_matrix()
```

```python
X.shape
```

```python
np.unique(X[:,2])
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
for c in [0.03125]:
    tmp = X[X[:,1] == c]
    print(tmp)
    plt.plot(tmp[:,2], 1 - tmp[:,0])

plt.xlabel("PCA components")
plt.ylabel("Error value")
plt.title("Error vs PCA components for C = 0.03125 \n(Linear SVM for DNA Methylation)")
```

```python
fig25 = plt.figure(figsize=(10,5))
ax25 = fig25.add_subplot(111)
for c in [20]:
    tmp = X[X[:,2] == c]
    plt.plot(tmp[:,1], 1 - tmp[:,0])

plt.xlabel("PCA components")
plt.ylabel("Error value")
plt.title("Error vs C for PCA 4 \n(Linear SVM for DNA Methylation)")
```

```python
X_axes = ['L_svm', 'G_svm', 'AdaBoost', 'Logistic Regression', 'G. Naive Bayes']
heights = np.array([85.71, 93.51, 90.90, 92.21, 81.82])
runtimes = np.array([4, 15, 2, 5, 1])
#colors = ['#C95834','#728CAC','#AA8D66','#319E98','#AF69AF']

ind = np.arange(len(X_axes))
runtime_color = '#AA8D66'
accuracy_color = '#319E98'

fig2 = plt.figure(figsize=(12,5))
ax2 = fig2.add_subplot(111)
ax4 = ax2.twinx()

ax2.bar(ind+0.125, heights-80, width=0.25, color=accuracy_color, bottom=80)
ax2.tick_params(axis='y', colors=accuracy_color)
ax2.set_ylabel('Accuracy (%)', color=accuracy_color)
ax2.spines['left'].set_color(accuracy_color)

ax4.bar(ind+0.375, runtimes, width=0.25, color=runtime_color)
ax4.spines['right'].set_color(runtime_color)
ax4.tick_params(axis='y', colors=runtime_color)
ax4.set_ylabel('Runtime (mins)', color=runtime_color)

runtime_proxy = plt.Rectangle((0, 0), 1, 1, color=runtime_color)
accuracy_proxy = plt.Rectangle((0, 0), 1, 1, color=accuracy_color)

#ax5.legend([runtime_proxy,accuracy_proxy],['Runtime','Accuracy'], loc='upper center')

plt.xticks(ind+0.25, X_axes, rotation=0)
ax2.set_xlabel("Classifiers")
fig2.suptitle("Performance of various classifiers for Protein Expression data")
```
